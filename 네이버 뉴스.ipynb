```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001'
html=requests.get(url).text
soup=BeautifulSoup(html,"html.parser")
All=soup.find('ul',{'class':'type06_headline'})#얘는 딱히 없어도 상관은 없음
```

```python
dl=All.find_all('dl')
for item2 in dl:
    try:
        img=item2.find('dt',{'class':'photo'}).find("img") #img src로 시작하기 때문
        print('img=',img['src']) #src=하이퍼링크형태로 가져오기
        print('----------------------------')
    except:
        print('no image')
```

```python
dl=All.find_all('dl')
for item2 in dl:
    try:
        img=item2.find('dt',{'class':'photo'}).find("img")['src'] #img src로 시작하기 때문
        print('img=',img) #src=하이퍼링크형태로 가져오기
        print('----------------------------')
    except:
        print('no image')
```

```python
#타이틀만 뽑아오기
dl=All.find_all('dl')
for item2 in dl:
    link=item2.find('dt',{'class':''}).find('a')
    print(link['href'])
    print(link.text.replace('\t','').replace('\n','')[1:len(link.text)+1])
```
