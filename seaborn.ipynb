```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#가지고 있는 데이터set 받아서 사용 가능(미리 있는게 있음). matplot은 random으로 발생시킴
tips=sns.load_dataset('tips') #데이터프레임형태
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
tips=sns.load_dataset('tips') #데이터프레임형태
sns.relplot(x='total_bill',y='tip',hue='sex',data=tips, style='time') 
#hue=scatter 변수들 색으로 구분 하는거, 단 categorial 변수여야함
#style도 바꾸면 --> 그 변수도 모양으로 구분해줌
```

```python
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g=sns.relplot(x='time',y='value',kind='line',data=df,color='green')
g.fig.autofmt_xdate()
```

```python
flights_long=sns.load_dataset('flights')
#month=index,year=column,승객수=value잡으려고 pivot씀
flights=flights_long.pivot('month','year','passengers')
f,ax=plt.subplots(figsize=(9,6))

sns.heatmap(flights,annot=True,fmt='d',linewidths=0.5,ax=ax)
```

```python
sns.set(style='white')
mpg=sns.load_dataset('mpg')
sns.relplot(x='horsepower',y='mpg',hue='origin',size='weight',sizes=(40,400),alpha=0.6,palette='muted',
          height=6,data=mpg)
```
